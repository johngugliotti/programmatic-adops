```python
import random,json

configs = []
def floors(config):
    for b in config.get('site'):
        for d in config.get('device'):
            for o in config.get('os').get(d):
                for s in config.get('slot').get(d):
                    for l in config.get('location'):
                        discount = 30
                        if o == 'ios': discount += 50
                        if l != 'us': discount += 25
                        hival = random.random() > .88
                        hival_factor = 4 if hival else 1
                        floor = (hival_factor) * (random.randint(0,200-discount)*.05)
                        cpi = floor/1000
                        #print(b,l,d,o,s,round(floor,2),round(cpi,6))
                        slot_params = {
                            "site": b
                            ,"geo": l
                            ,"device": d
                            ,"os": o
                            ,"slot": s
                            ,"floor": round(floor,2)
                            ,"price_level": "high" if hival else "norm"
                            
                        }
                        configs.append(slot_params)
    return configs                    

slot_configs = floors(config)
print(json.dumps(slot_configs,indent=2))




    
```

```python
def sim_slot_level_metrics(slot_params={}, start_date=datetime, end_date=datetime, anomalies=[]):
    upside = .5 if slot_params.get("geo") == "us" and slot_params.get("os") != "ios" else .1
    baseline_cpm = slot_params.get('floor') * ( 1 + random.random()* upside)
    percent_above_floor=.05
    #print(percent_above_floor)
    #baseline_cpm =  slot_params.get('floor') + percent_above_floor
    timespan_days = (end_date - start_date).days * 24
    
    return np.random.normal(loc=baseline_cpm, scale=  percent_above_floor, size=timespan_days)
    
    
```
